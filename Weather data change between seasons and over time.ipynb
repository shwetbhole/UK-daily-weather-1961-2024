Importing Libraries

```python
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')

import warnings
warnings.filterwarnings("ignore")
```

```python
weather_data = pd.read_csv("land_uk_daily_v2.csv")
weather_data
```

```python
weather_data['season'] = weather_data['month'].replace({
    1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring',
    5: 'Spring', 6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Autumn', 10: 'Autumn', 11: 'Autumn', 12: 'Winter'
})

weather_data
```

Summary statistics for temperature grouped by seasaon

You can calculate the statistics for the temperature grouped by season using the code

```python
weather_data.groupby('season')['temp'].describe().round(2)
```

```python
sns.catplot(data=weather_data, x='season', y = 'temp', kind='box')
```

```python
weather_data.groupby('month')['temp'].describe().round(2)
```

```python
sns.catplot(data=weather_data, x='month', y='temp', kind='box');
```

```python
sns.catplot(data=weather_data, x='month', y='temp', kind='bar');
```

```python
weather_data.groupby('season')['wind_speed'].describe().round(2)
```

```python
sns.catplot(data=weather_data, x='month', y='wind_speed', kind='box');
```

```python
sns.boxplot(data=weather_data, x= 'month', y='wind_speed')
```

```python
weather_data.groupby('season')['precipitation'].describe().round(2
                                                                 )
```

```python
sns.catplot(data=weather_data, x='season', y='precipitation', kind='box');
```

```python
weather_data.groupby('month')['precipitation'].describe().round(2)
```

```python
sns.catplot(data=weather_data, x='month', y= 'precipitation', kind= 'box');
```
